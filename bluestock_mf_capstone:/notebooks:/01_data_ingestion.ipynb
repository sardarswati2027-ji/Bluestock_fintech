LOAD ALL CSV FILES FROM GOOGLE DRIVE

```python
import pandas as pd
from google.colab import drive
import os

drive.mount('/content/drive')
```

```python
folder_path = "/content/drive/MyDrive/Colab Notebooks/Bluestock_MF_Datasets"

csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

print(f"Total CSV files found: {len(csv_files)}\n")

datasets = {}

for file in sorted(csv_files):

    file_path = os.path.join(folder_path, file)

    print("=" * 70)
    print(f"DATASET: {file}")
    print("=" * 70)

    df = pd.read_csv(file_path)

    datasets[file] = df

    print("\nShape:")
    print(df.shape)

    print("\nData Types:")
    print(df.dtypes)

    print("\nFirst 5 Rows:")
    print(df.head())

    print("\n\n")
```
