```python
import requests
import pandas as pd
from google.colab import drive
import os

drive.mount('/content/drive')

url = "https://api.mfapi.in/mf/125497"

response = requests.get(url)

data = response.json()
```

Extract NAV history data

```python
nav_data = data['data']

df = pd.DataFrame(nav_data)

print("First 5 rows:")
print(df.head())

print("\nShape:")
print(df.shape)
```

Saving in Google Drive

```python
save_folder = "/content/drive/MyDrive/Bluestock_MF_Datasets"

os.makedirs(save_folder, exist_ok=True)

csv_path = os.path.join(save_folder, "live_nav_hdfc_top100.csv")

df.to_csv(csv_path, index=False)

print(f"\nCSV saved successfully at:\n{csv_path}")
```
